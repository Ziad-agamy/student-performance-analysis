### Inferential Statistics: Chi-squared Test for Independence

The code block below performs a Chi-squared test for independence to determine if there is a statistically significant association between various categorical variables and the `GPA_Class`. The test assesses whether the distribution of `GPA_Class` is independent of the distribution of each categorical feature. A `p-value less than 0.05` indicates a statistically significant dependency.

```python
cat_cols = ["Race", "ParentalEducation", "SES_Quartile", "Locale", "SchoolType", "Gender",
               "InternetAccess", "Extracurricular", "PartTimeJob", "ParentSupport", "Romantic", "FreeTime", "GoOut"]
dependent = []
for col in cat_cols:
  test_corr = pd.crosstab(df["GPA_Class"], df[col])
  chi2_result = stats.chi2_contingency(test_corr)
  if chi2_result[1] < 0.05:
    dependent.append(col)
print(f"features that has relationship associated with GPA Score: \n{dependent}")
```

```python
cols = ["ParentalEducation", "Race", "SES_Quartile", "SchoolType", "Locale", "ParentSupport"]
for col in cols:
    gpa = df.groupby([col])["GPA"].mean()
    print(gpa)  
    print("\n")
```

### Investigating Outliers

The code block below filters outliers based on `Study Hours` to prepare for further analysis and Investigating if there is any key differences between the sample dataset and outliers

```python
q1 = df["StudyHours"].quantile(0.25)
q3 = df["StudyHours"].quantile(0.75)
iqr = q3 - q1

studyhours_outliers_upp = df[df["StudyHours"] > q3 + (1.5*iqr)]
studyhours_outliers_low = df[df["StudyHours"] < q1 - (1.5*iqr)]
```

### Confidence Intervals

The code block below calculates the 99% confidence intervals for the mean `Attendance Rate` and mean `Study Hours` for students categorized as having an `"Excellent"` GPA. A confidence interval provides a range of values within which the true population mean is likely to lie, based on the sample data.

```python
more_3gpa = df[df["GPA_Class"] == "Excellent"]
int_test = ["AttendanceRate", "StudyHours"]

for i, col in enumerate(int_test):
  x_bar = more_3gpa[col].mean()
  n = more_3gpa[col].count()
  s = more_3gpa[col].std()

  se = s / np.sqrt(n)
  intervals = stats.norm.interval(0.99, loc=x_bar, scale=se)

  print(f"Mean {col} for students with Excellent Degree: {x_bar:.3f}")
  print(f"Based on the analysis, with 99% Confidence, the mean of {col} for students, given an Excellent Degree, lies between {intervals[0]:.4f} and {intervals[1]:.4f}.")
  print("\n") if i == 0 else None
```

### Comparing Groups: Independent Samples t-tests

This section performs independent samples `t-tests` to compare the means of `'AttendanceRate'` and `'StudyHours'` between students with `"Excellent"` and `"Very Good"` GPA classes. The t-test determines if the observed difference in means between the two groups is statistically significant. A low p-value (typically < 0.05) suggests that the difference is unlikely to be due to random chance.

```python
for i, col in enumerate(int_test):
  excellent = df[df["GPA_Class"] == "Excellent"][col]
  very_good = df[df["GPA_Class"] == "Very Good"][col]

  test_results = stats.ttest_ind(excellent, very_good)

  if test_results[1] < 0.05:
    print(f"Rejected the null with p-value: {test_results[1]}")
    print(f"There is statistically significant difference in {col} between 'Excellent' and 'Very Good' GPA students")
  else:
    print(f"Failed to reject the null with p-value: {test_results[1]}")
    print(f"There is not enough evidence to conclude that there is a statistically significant difference in {col} between 'Excellent' and 'Very Good' GPA students")
  print("\n") if i == 0 else None
```

## Regression Modeling

This section focuses on building a linear regression model to predict a student's GPA based on key numerical features: `Total Test Score`, `Study Hours`, and `Attendance Rate`. The data is first split into training and testing sets to train the model on a subset of the data and then evaluate its performance on unseen data.

```python
df = df.sample(frac=1, random_state=42).reset_index(drop=True)
predictors = [
    "TotalScore",
    "StudyHours",
    "AttendanceRate"
]

X = sm.add_constant(df[predictors])
y = df["GPA"]

X_train = X.iloc[:10000]
X_test = X.iloc[10000:]

y_train = y.iloc[:10000]
y_test = y.iloc[10000:]
```

### Model Training and Evaluation

A linear regression model is trained using the Ordinary Least Squares (OLS) method on the training data. The model's performance is then evaluated by predicting GPA on the test set and comparing these predictions to the actual GPA values.

```python
lm = sm.OLS(y_train, X_train)
results = lm.fit()
print(results.summary())
```

```python
y_hat = results.predict(X_test)
y_actual = y_test

plt.figure(figsize=(10, 6))
ax = sns.regplot(x=y_hat, y=y_actual, line_kws={'color': 'black'}, color='darkcyan')
plt.xlabel("Predicted GPA")
plt.ylabel("Actual GPA")
plt.grid(linestyle='--')
plt.show()
```

### Model Performance Metrics

To quantitatively assess the performance of the linear regression model, we calculate the Mean Squared Error (MSE) and Root Mean Squared Error (RMSE). MSE represents the average squared difference between the predicted and actual GPA values, while RMSE is the square root of MSE, providing a measure of the typical prediction error in the same units as GPA. Lower values for both MSE and RMSE indicate better model accuracy.

```python
rows, cols = X_test.shape
residuals = y_actual - y_hat

mse = (1/rows) * sum(residuals**2)
rmse = np.sqrt(mse)

print(f"Mean Square Error: {mse:.4f}")
print(f"Root Mean Square Error: {rmse:.4f}")
```

## Conclusion

This analysis explored various factors influencing student performance, particularly Grade Point Average (GPA), using a comprehensive dataset.

The exploratory data analysis revealed the distributions of demographic factors, academic indicators (test scores, attendance, study hours), and other student-related variables. While initial chi-squared tests did not show statistically significant dependencies between categorical variables (like Race, Parental Education, etc.) and the classified 'GPA_Class', examining the mean GPA across categories and proportional distributions provided some visual insights, though subtle. For instance, students with Bachelor's+ or HS parental education had slightly higher mean GPAs. Students with internet access and those involved in extracurricular activities showed slightly higher proportions in the "Excellent" and "Very Good" GPA classes.

Correlation analysis highlighted strong positive correlations between test scores (Math, Reading, Science, and Total Score) and GPA, as well as between Attendance Rate and GPA. Study Hours showed a moderate positive correlation with GPA. These relationships were further visualized through scatter plots, confirming the positive trends. Box plots of Attendance Rate and Study Hours across GPA classes showed increasing median values as GPA class improved from "Fail" to "Excellent", and a t-test confirmed statistically significant differences in the means of these factors between "Excellent" and "Very Good" GPA students. Analysis of study hours outliers suggested that both very low and very high study hours might be associated with lower GPA classes, and Parental Education distribution among these outliers also showed some variations compared to the overall distribution.

A linear regression model was built to predict GPA using Total Test Score, Study Hours, and Attendance Rate as predictors. The model achieved an R-squared value of 0.584, indicating that approximately 58.4% of the variance in GPA can be explained by these variables. The Mean Squared Error (MSE) of 0.0908 and Root Mean Squared Error (RMSE) of 0.3014 on the test set suggest that the model's predictions are reasonably close to the actual GPA values, with a typical prediction error of around 0.3 GPA points. The regression results indicated that Total Score is a highly significant predictor of GPA, while Study Hours and Attendance Rate, though positively correlated, were not statistically significant predictors *when considered alongside Total Score in this specific linear model*. This could imply that their impact on GPA might be largely captured by their influence on test scores and overall academic performance.

In conclusion, this analysis confirms that academic performance measures like test scores, attendance rate, and study hours are significant factors associated with GPA. While categorical factors did not show strong statistical independence in the chi-squared tests with the classified GPA groups, some subtle trends were observed. The linear regression model effectively predicts GPA based on Total Test Score, Study Hours, and Attendance Rate, with Total Test Score being the most prominent predictor in the model. Future analysis could explore non-linear relationships, interactions between variables, and incorporate categorical features through appropriate encoding to potentially improve the model and gain deeper insights into the complex interplay of factors influencing student performance.
